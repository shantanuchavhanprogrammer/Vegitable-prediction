```python
import pandas as pd
```

```python
import os
```

```python
import glob
```

```python
import cv2
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img= cv2.imread(os.path.join(os.getcwd(), "Vegetable Images", "Bean","0001.jpg"))
```

```python
img.shape
```

```python
image_size=(64,64)
```

```python
data_generator= ImageDataGenerator(
    rotation_range=20
    rescale=1/255, validation_split=0.2)
```

```python
train_generator= data_generator.flow_from_directory('Vegetable Images/', target_size=image_size, batch_size=32, class_mode="sparse", subset='training')
```

```python
val_generator = data_generator.flow_from_directory("Vegetable Images/", target_size=image_size, batch_size=32, class_mode="sparse", subset='validation')
```

```python
from tensorflow.keras import Sequential
```

```python
model= Sequential()
```

```python
from tensorflow.keras.layers import Conv2D
```

```python
model.add(Conv2D(filters=16,kernel_size=(3,3), strides=(1,1), activation='relu', input_shape=image_size +(3,)))
```

```python
from tensorflow.keras.layers import AveragePooling2D
```

```python
model.add(AveragePooling2D(pool_size=(2,2)))
```

```python
from tensorflow.keras.layers import Flatten
```

```python
model.add(Flatten())
```

```python
from tensorflow.keras.layers import Dense
```

```python
model.add(Dense(units=16, activation='relu'))
```

```python
model.add(Dense(units=15, activation='softmax'))
```

```python
model.compile(optimizer="adam", loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop= EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)
```

```python
history= model.fit(train_generator, validation_data=val_generator, callbacks=[early_stop], epochs=20, verbose=1)
```
